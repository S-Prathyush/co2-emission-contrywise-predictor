```python
# Target CO2 emission column names
co2_cols = ['Metric tons of CO2e per capita (1990)',
            'Metric tons of CO2e per capita (1995)',
            'Metric tons of CO2e per capita (2000)',  # Add new columns
            'Metric tons of CO2e per capita (2005)',
            'Metric tons of CO2e per capita (2013)',
            'Metric tons of CO2e per capita (2018)']

# Calculate average for each CO2 emission column (handling non-numeric data)
avg_co2_values = {}
for col in co2_cols:
  try:
    # Attempt conversion to numeric, calculate average if successful
    avg_co2_values[col] = df[col].astype(float).mean()  # Try numeric conversion
  except (ValueError, TypeError):
    print(f"Warning: Column '{col}' contains non-numeric data. Skipping average calculation.")

# Fill missing values in each CO2 emission column with its own average (if available)
for col, avg_value in avg_co2_values.items():
  if avg_value is not None:  # Check if average was calculated
    df[col].fillna(avg_value, inplace=True)
```

```python
df.isnull().sum()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Create encoders for categorical columns
le_country = LabelEncoder()
le_continent = LabelEncoder()
le_hemisphere = LabelEncoder()

# Encode each categorical column and add a new column with encoded values
df['Country'] = le_country.fit_transform(df['Country'])
df['Continent'] = le_continent.fit_transform(df['Continent'])
df['Hemisphere'] = le_hemisphere.fit_transform(df['Hemisphere'])
df
```

```python
# Remove the "ISO3" column
df = df.drop('ISO3', axis=1)  # axis=1 specifies removing a column
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Feature selection (predictors) - Choose relevant columns for prediction
features = [col for col in df.columns if col not in ['Metric tons of CO2e per capita (2000)']]  # Exclude target

# Target variable (what we want to predict)
target = 'Metric tons of CO2e per capita (2013)'  # Choose the target year

# Encode categorical features (optional, but recommended for some models)
categorical_cols = ['Country', 'Continent', 'Hemisphere']  # Assuming encoded columns
le = LabelEncoder()
for col in categorical_cols:
  df[col] = le.fit_transform(df[col])  # Encode each categorical column (optional)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Create the Random Forest Regression model
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the testing set
y_predicted = model.predict(X_test)

# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_predicted)
r2 = r2_score(y_test, y_predicted)

# Print the evaluation metrics
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R^2): {r2:.2f}")
```

```python

# Example prediction for a specific country number and future year (replace with desired values)
country_number = 123  # Replace with the actual country number
year_to_predict = 2081 # Replace with desired year (assuming data exists for relevant features)

# Filter data for the specific country (assuming country number as index)
try:
  country_data = df.iloc[country_number]  # Access data using country number as index
except IndexError:
  print(f"Country number {country_number} is out of range.")
  exit()
# Check if data exists and features are available
if set(features).issubset(country_data.index):
  # Prepare features for prediction (consider data transformation if needed)
  prediction_features = country_data[features].values.reshape(1, -1)  # Reshape for single prediction

  # Make prediction for the specific country and year
  predicted_co2 = model.predict(prediction_features)[0]
  print(f"Predicted CO2 emission for country number {country_number} in {year_to_predict}: {predicted_co2:.2f} tons per capita")
else:
  print(f"Data for country number {country_number} or required features might be missing for prediction in {year_to_predict}.")

```
